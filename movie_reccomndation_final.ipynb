The collumns that we need for our reccomendation systms are
movie_id,cast,crew,genres,keywords,title

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head()
# the collum are list of dict we have to bring it in list format
# [action,adventure,fantasy]
```

```python
import ast
# string indices must be integers, not 'str'
# to resolve this error we use ast 
# ast.litral is used to convert stings of list inside dict into list
```

```python
# convert function is used to convert the dict into list
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
# .apply method is used to apply a function along the axis of a DataFrame or Series
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
# simillar step
```

```python
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
# we only need top 3 actors name 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
# from the dict where job has director 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.sample(5)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
# we have to remove the space between names in collum so that name and sirname are not treated as diffrent tags
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
# The lambda function seems to split each string in the 'overview' column into a list of words
# because every other collumn is list and and to merge it with others it neend to be split in list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
# make a new collumn tags to join all the other function into tags
```

```python
# 'new' that is a copy of the original 'movies' DataFrame, but without the specified columns
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
new.tags[[0]]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# it is used for converting a collection of text documents to a matrix of token counts
cv = CountVectorizer(max_features=5000,stop_words='english')
# max_features specify unique words or terms. the vectorizer will consider the top 5000 most frequent words in this corpus
```

```python
vector = cv.fit_transform(new['tags']).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#for reccomending we will find cosin distance i.e angel 
```

```python
similarity = cosine_similarity(vector)
similarity
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
# we use ennum function so that we can hold to the index 
```

```python
recommend('Gandhi')
```
